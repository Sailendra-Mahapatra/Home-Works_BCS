### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime


# Import API key
from config import Weather_API

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
# output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

url = 'http://api.openweathermap.org/data/2.5/weather?'

# Save responses

coord1_lat=[]
coord2_lon=[]
tempr_max=[]
wind=[]
humid_percent=[]
cloud=[]
cityname=[]
city_country=[]
weather_date=[]
counter_set=1
counter_rec=0

    
```

```python
#print (cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python

print('Beginning Data Retrieval ')
print('--http://localhost:8825/notebooks/Weather_V9.ipynb#-------------------------')


for city_name in cities:
    city_select = url + 'appid='+ Weather_API + '&q=' + city_name + '&units=imperial'
    resp = requests.get (city_select)
    wresp = resp.json()


# printing Data retrieval records in set of 50 cities

    counter_rec=counter_rec+1
    if counter_rec>50:
        counter_set=counter_set+1
        counter_rec=1
        
    
    
    
    if resp.status_code ==200:
        print(f'Processing Record {counter_rec} of Set {counter_set} | {city_name}')
        
        coord1_lat.append(wresp['coord']['lat'])
        coord2_lon.append(wresp['coord']['lon'])
        tempr_max.append(wresp['main']['temp_max'])
        humid_percent.append(wresp['main']['humidity'])
        wind.append(wresp['wind']['speed'])
        cloud.append(wresp['clouds']['all'])
        cityname.append(wresp['name'])
        city_country.append(wresp['sys']['country'])
        weather_date.append(wresp['dt'])
        
        
    else:
        print(f'City not found.  Skipping ....')


                      
    
    
    
    
```

```python
#dict1={resp:'resp_list',tempr:'temp_resp',tempr',humidity_percent,cloudiness_percent,wind_speed_mph}

weather = {'City Name': cityname,
                'Country':city_country,
                'Date':weather_date,
                'Latitude':coord1_lat,
                'Longitude':coord2_lon, 
                'Max Temperature in F': tempr_max, 
                'Humidity (%)': humid_percent,
                'Wind Speed in MPH': wind,
                'Cloudness (%)': cloud
              }

# create data frame as per above columns
#print(weather)
df = pd.DataFrame(weather)
df.head(5)
```

```python
date_now=datetime.fromtimestamp(df['Date'].max()).strftime('%Y/%m/%d')
```

#### Latitude vs. Temperature Plot

```python
# Scatter Plot generations
# Latitude vs Max Temp

plt.figure(figsize=(10,7))
plt.scatter(df['Latitude'],df['Max Temperature in F'],
            s=50,marker='o',color='grey', edgecolor='black')

plt.title('City Latitude vs Max Temperature ('+date_now+')',fontsize=18)
plt.xlabel('Latitude',fontsize=14)
plt.xlim(-90,90)
plt.ylabel('Max Temperature in F',fontsize=14)
plt.grid(True)
plt.savefig('HW-6_Latitude-Max Temperature Plot')
plt.show()
```

#### Latitude vs. Humidity Plot

```python
# Latitude vs Humidity

plt.figure(figsize=(10,7))
plt.scatter(df['Latitude'],df['Humidity (%)'],
            s=50,marker='o',color='blue', edgecolor='black',alpha=.5)

plt.title('City Latitude vs Humidity (%)'' ('+date_now+')',fontsize=18)
plt.xlabel('Latitude',fontsize=14)
plt.ylim(0,110)
plt.ylabel('Humidity (%)',fontsize=14)
plt.grid(True)
plt.savefig('HW-6_Latitude-Humidity Plot.png')

plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
# Latitude vs Cloudness

plt.figure(figsize=(10,7))
plt.scatter(df['Latitude'],df['Cloudness (%)'],
            s=50,marker='o',color='grey', edgecolor='black',alpha=.8)

plt.title('City Latitude vs Cloudness (%)'' ('+date_now+')',fontsize=18)
plt.xlabel('Latitude',fontsize=14)
plt.xlim(-90,90)
plt.ylabel('Cloudness (%)',fontsize=14)
plt.grid(True)
plt.savefig('HW-6_Latitude-Cloudness Plot.png')

plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# Latitude vs Wind Speed

plt.figure(figsize=(10,7))
plt.scatter(df['Latitude'],df['Wind Speed in MPH'],
            s=50,marker='o',color='blue', edgecolor='black',alpha=.7)

plt.title('City Latitude vs Wind Speed'' ('+date_now+')',fontsize=18)
plt.xlabel('Latitude',fontsize=14)
plt.xlim(-90,90)
plt.ylabel('Wind Speed in MPH',fontsize=14)
plt.grid(True)
plt.savefig('HW-6_Latitude-Wind Speed Plot.png')

plt.show()
```
